```python
X= df[['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y= df['Species']
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

# Logistic Regression

```python

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

# Train the model on our train dataset
lr.fit(X,y)

# Train the model with the training set

lr.fit(X_train,y_train)
```

```python
predictions = lr.predict(X)

# Compare with the actual charges

Scores = pd.DataFrame({'Actual':y,'Predictions':predictions})
Scores.head()
```

```python
y_test_hat=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_test_hat)*100,'%')
```

# SVM

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
svm.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
svm_pred = svm.predict(X_test)
clf_report_svm = pd.DataFrame(classification_report(y_test, svm_pred, output_dict=True))
clf_report_svm
```

```python
svm_test_hat=svm.predict(X_test)
```

```python
print(accuracy_score(y_test_hat,svm_test_hat)*100,'%')
```
